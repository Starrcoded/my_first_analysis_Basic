```python
# i cant figure out why it brought out none', I wanted to use Sorte by rating not using value counts
jumia_project_sorted = jumia_project.sort_values(by=['current_price'], inplace=True)

print(jumia_project_sorted)
```

```python
# To get actual number of each brand, I used value counts, 
# this will help me reduce or extract the files to work with and eliminate the less frequent brands 
# and having less data to compare with.

jumia_project['brand'].value_counts()
```

```python
# Here I used the value counts and .index to eliminate less frequent brands with least counts
# so i picked counts > 3
# so I am left with brands with counts > 4 to work on.

high_brands=jumia_project['brand'].value_counts()[jumia_project['brand'].value_counts() > 3].index
high_brands
```

```python
# extracting the high brands based on value count above

best_highbrands= jumia_project[jumia_project['brand'].isin(high_brands)]
best_highbrands
```

```python
# from the mean and median of this set of data, i can consider BEST (TOP) LOW Phones of approximately 54,000 and 
# 73,000, both cameras are of higher pixel, battery life higher than 4000, android version more than 8

best_highbrands.mean()
best_highbrands.median()
print(best_highbrands.mean())


print(best_highbrands.median())
```

```python
# so I would recomend these sets of top low cost phones to my Client,
# Considering the current price <53900 which is the meadian of the set.

top_lowcost = best_highbrands.loc[best_highbrands['current_price'] <= 53900]
top_lowcost.head() 
```
